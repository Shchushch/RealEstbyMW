# Этажность

### Добавим колонку "этаж квартиры" и "этажность"

```python
rlst['Дом'].apply(lambda x: x.split(',')[0]).unique()
```

```python
rlst['Этаж квартиры'] = [int(row.split('/')[0]) for row in rlst['Дом']]
rlst['Этажность дома'] = [int(row.split('/')[1].split(', ')[0]) for row in rlst['Дом']]
rlst
```

```python
floors = pd.DataFrame(rlst['Этажность дома'].value_counts(), columns=['Этажность дома'])
floors.head(30) #видим, как много пятиэтажек
```

```python
five_floors = rlst.loc[:, ['Этажность дома', 'Мусоропровод']]
five_floors = five_floors[five_floors['Этажность дома'] == 5]
five_floors #понадобится в анализе "Мусоропровода"
```

```python
def typee(row) :
    result = {}
    row = row['Дом'].split(',')
    if len(row) == 0:
        return result
    if len(row) > 1:
        result['building_type'] = row[1][1:]
    return result
```

```python
house_type = rlst.apply(typee, axis=1, result_type='expand') #определили типы домов и самый частый - панельный
house_type['building_type'].value_counts()
```

```python
rlst = rlst.drop('Дом', axis = 1)
```

```python
rlst['Тип дома'] = house_type['building_type']
rlst['Тип дома'] = rlst['Тип дома'].fillna('Панельный')
```

# Парковка

```python
rlst['Парковка'] = rlst['Парковка'].fillna(0) #NaN приравниваем к 0, ответа 'нет' в датасете не найдено
```

```python
rlst['Парковка'] = rlst['Парковка'].replace(['наземная'], 1)
rlst['Парковка'] = rlst['Парковка'].replace(['подземная'], 1)
rlst['Парковка'] = rlst['Парковка'].replace(['открытая'], 1)
rlst['Парковка'] = rlst['Парковка'].replace(['многоуровневая'], 1)
rlst['Парковка'] = rlst['Парковка'].replace(['на крыше'], 1)
```

# Балкон

```python
rlst['Балкон'].value_counts() # NaN будем приравнивать к отсутствию балкона/лоджии
```

```python
rlst['Балкон'] = rlst['Балкон'].fillna('Балкон (0)')
rlst['Лоджии'] = [x.split(' ')[-1] if x.split(' ')[-2] == 'Лоджия' else '(0)' for x in rlst['Балкон']]
rlst['Лоджии'] = [int(x.replace('(', '').replace(')', '')) for x in rlst['Лоджии']]
```

```python
rlst['Балконы'] = [x.split(' ')[1] if x.split(' ')[0] == 'Балкон' else '(0)' for x in rlst['Балкон']]
rlst['Балконы'] = [int(x.replace('(', '').replace(')', '').replace(',', '')) for x in rlst['Балконы']]
```

# Санузел

```python
rlst['Санузел'].value_counts() #NaN будем приравнивать к одному совмещенному санузлу
```

```python
rlst['Санузел'] = rlst['Санузел'].fillna('Совмещенный (1)')
```

```python
#выделим цифры
ints = rlst['Санузел'].str.findall(r'(\d+(?:\.\d+)?)').values #regex
ints
```

```python
temp = []
for i in range(len(ints)):
    ints[i] = [int(x) for x in ints[i]]
    temp.append(sum(ints[i]))
rlst["Количество санузлов"] = [x for x in temp]
```
