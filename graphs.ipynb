```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('_data.csv', index_col = [0])
```

```python
df 
```

```python
df['Адрес'].apply(lambda x: x.split(',')[0]).unique()
```

```python
df.drop(df[df['Адрес'].str.startswith('Санкт-Петербург')].index, inplace = True)
df.drop(df[df['Адрес'].str.startswith('Краснодарский край')].index, inplace = True)
df #рассматриваем только квартиры в мск
```

```python
df['Адрес'].apply(lambda x: x.split(',')[0]).unique()
```

```python
summary1 = pd.concat([df.count(), df.isnull().sum()], axis = 1).reset_index()
summary1.rename(columns = {'index':'параметр', 0:'ненулевых', 1:'нулевых'}, inplace = True)
summary1
```

```python
summary1.set_index('параметр', inplace = True)
plt.barh(y = summary1.index, width = summary1['ненулевых'], left=summary1['нулевых'])
plt.barh(y = summary1.index, width = summary1['нулевых'])
#оранжевые - нулевые данные
#синие - ненулевые
```

```python
summary2 = pd.concat([df.nunique(), df.count()], axis = 1).reset_index()
summary2.rename(columns = {'index':'параметр', 0:'уникальный', 1:'всего'}, inplace = True)
summary2
```

```python
sns.relplot(data=summary2, kind = 'line')
```
