```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

rlst = pd.read_csv('_data.csv', index_col = [0])
```

```python
rlst 
```

```python
rlst['Адрес'].apply(lambda x: x.split(',')[0]).unique()
```

```python
rlst.drop(rlst[rlst['Адрес'].str.startswith('Санкт-Петербург')].index, inplace = True)
rlst.drop(rlst[rlst['Адрес'].str.startswith('Краснодарский край')].index, inplace = True)
rlst #рассматриваем только квартиры в мск
```

```python
rlst['Адрес'].apply(lambda x: x.split(',')[0]).unique()
```

```python
Nulls = pd.concat([rlst.count(), rlst.isnull().sum()], axis = 1).reset_index()
Nulls.rename(columns = {'index':'параметр', 0:'ненулевых', 1:'нулевых'}, inplace = True)
Nulls
```

```python
Nulls.set_index('параметр', inplace = True)
plt.barh(y = Nulls.index, width = Nulls['ненулевых'], left=Nulls['нулевых'])
plt.barh(y = Nulls.index, width = Nulls['нулевых'])
#оранжевые - нулевые данные
#синие - ненулевые
```

```python
summary2 = pd.concat([rlst.nunique(), rlst.count()], axis = 1).reset_index()
summary2.rename(columns = {'index':'параметр', 0:'уникальный', 1:'всего'}, inplace = True)
summary2
```

```python
sns.relplot(data=summary2, kind = 'line')
```
